```python
import pandas

df = pandas.read_csv('trapezoid.csv')
df.drop(labels="cmd", axis=1, inplace=True)
print(df)
axes = df.plot.line(x="time", subplots=True, layout=(2,2))
print(axes[0][0])
```

```python
import matplotlib.pyplot as plt
import numpy as np

def make_plot(axs):
    box = dict(facecolor='yellow', pad=5, alpha=0.2)

    # Fixing random state for reproducibility
    np.random.seed(19680801)
    ax1 = axs[0, 0]
    ax1.plot(df["time"], df["leftPWM"])
    ax1.set_title('Left Motor')
    ax1.set_ylabel('PWM')
    ax1.set_ylim(0, 255)

    ax3 = axs[1, 0]
    ax3.set_ylabel('Encoder Counts')
    ax3.plot(df["time"], df["leftEnc"], 'r')
    ax3.set_ylim(0, 3000)

    ax2 = axs[0, 1]
    ax2.set_title('Right Motor')
    ax2.plot(df["time"], df["rightPWM"])
    ax2.set_ylim(0, 255)

    ax4 = axs[1, 1]
    ax4.plot(df["time"], df["rightEnc"], 'r')
    ax4.set_ylim(0, 3000)


# Plot 1:
fig, axs = plt.subplots(2, 2)
#fig.subplots_adjust(left=0.2, wspace=0.6)
make_plot(axs)

# just align the last column of axes:
plt.show()
```
